```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

from sklearn.decomposition import FastICA, PCA
```

```python
#data=np.load('2019_05_05_14_03_59_bindedimages.npy')
data=np.load('2019_05_05_13_43_01_bindedimages.npy')
#data=np.load('2019_05_05_13_53_27_bindedimages.npy')
plt.matshow(data[101])
reshaped=data.reshape(data.shape[0],(data.shape[1]*data.shape[2])).astype(float)
reshaped.shape
# Compute ICA

ica = FastICA(n_components=121,max_iter=2000)
ica.fit(reshaped)
A_ica=ica.components_
image_sum=np.sum(np.abs(A_ica),axis=0).reshape(data.shape[1],data.shape[2])
plt.matshow(image_sum)
```

```python
from PIL import Image
img=Image.fromarray(np.asarray(image_sum))
```

```python
rotated=img.rotate(1)
```
