```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

from sklearn.decomposition import FastICA, PCA
```

```python
#data=np.load('2019_05_05_14_03_59_bindedimages.npy')
data=np.load('2019_05_05_13_43_01_bindedimages.npy')
data2=np.load('20190814_bindedimages.npy')
#data=np.load('2019_05_05_13_53_27_bindedimages.npy')
plt.matshow(data[101])
reshaped=data.reshape(data.shape[0],(data.shape[1]*data.shape[2])).astype(float)
reshaped.shape
# Compute ICA

ica = FastICA(n_components=121,max_iter=2000)
ica.fit(reshaped)
A_ica=ica.components_
image_sum=np.sum(np.abs(A_ica),axis=0).reshape(data.shape[1],data.shape[2])
plt.matshow(image_sum)
```

```python
plt.matshow(image_sum)
```

```python
import scipy.ndimage

normalizedimage=image_sum/np.max(image_sum)
rotated=scipy.ndimage.rotate(normalizedimage,-1.5)
#fig, ax = plt.subplots(figsize=(10,10))
fig=plt.figure()
fig.set_size_inches((5,5))
ax=plt.Axes(fig,[0,0,1,1])
ax.set_axis_off()
fig.add_axes(ax)
cm='Blues'
ax.imshow(rotated[5:62,18:75],cmap=cm,interpolation='none',vmin=0.12,vmax=1)
#fig.savefig(cm+'.pdf',box_inches='tight')
```
